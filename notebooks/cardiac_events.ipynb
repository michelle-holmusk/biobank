```python
def removenan(df): 
    datet2 = []
    diagt2 = []
    for i in range(df.shape[0]):
        temp1 = [x for x in df.iloc[i]['datet'] if type(x) != float]
        temp2 = [x for x in df.iloc[i]['diagt'] if type(x) != float]
        datet2.append(temp1)
        diagt2.append(temp2)
    df['datet2'] = datet2
    df['diagt2'] = diagt2

```

```python
def zipper(df):
    result = []
    for i in range(df.shape[0]):
        temp = list(map(list, zip(df.iloc[i]['datet2'], df.iloc[i]['diagt2'])))
        result.append(temp)
    df['zipped'] = result
    print('end')
```

```python
def icdsearch(df, code):
    result = []
    if code == 10:
        for x in df['zipped']:
            temp = []
            if len(x) > 0:
                for y in x:
                    q = y[1][:3]
                    if q in ICD[10]['all']:
                        diag = icd10_codes[icd10_codes['ICD10'].str.contains(q)]['diagnosis'].values[0]
                        temp.append([datetime.strptime(y[0], '%Y-%m-%d').strftime('%d/%m/%Y'), diag, 'sec'])
            result.append(temp)
    elif code == 9:
        for x in df['zipped']:
            temp = []
            if len(x) > 0:
                for y in x:
                    q = str(y[1])[:3]
                    if q in ICD[9]['all']:
                        diag = icd9_codes[icd9_codes['ICD9'].str.contains(q)]['diagnosis'].values[0]
                        temp.append([datetime.strptime(y[0], '%Y-%m-%d').strftime('%d/%m/%Y'), diag, 'sec'])
            result.append(temp)
    df['coded'] = result
    print('end')
    
```

```python
sort_record = []

for x in records['uniq_record_diag']:
    
    temp = []
    temp_pri = []
    temp_sec = []
    
    # handling nan
    for i in range(len(x)):
        if x[i] == [np.nan]:
            continue
        else: 
            #changing dates into datetime
            if type(x[i][0]) == str:
                yi0 = datetime.strptime(x[i][0], '%d/%m/%Y')
                yi1 = x[i][1]
                yi2 = x[i][2]
                temp.append([yi0, yi1, yi2])
                
    
    # sorting by dates
    temp.sort(key=lambda x: x[0])

    # append the sorted list
    sort_record.append(temp) 
```
